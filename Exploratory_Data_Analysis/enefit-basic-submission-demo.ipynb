```python
import enefit
import pandas as pd
env = enefit.make_env()
iter_test = env.iter_test()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df_client = pd.read_csv("/kaggle/input/predict-energy-behavior-of-prosumers/example_test_files/client.csv")
df_gas = pd.read_csv("/kaggle/input/predict-energy-behavior-of-prosumers/example_test_files/gas_prices.csv")
df_elec = pd.read_csv("/kaggle/input/predict-energy-behavior-of-prosumers/example_test_files/electricity_prices.csv")
```

```python
counter = 1
for (test, revealed_targets, client, historical_weather,
        forecast_weather, electricity_prices, gas_prices, sample_prediction) in iter_test:
    
    print("iteration N° : ",str(counter))
    print("Date min : ", test['prediction_datetime'].min())
    print("Row_id min test : ", test['row_id'].min())
    print("Row_id min sample prediction : ", sample_prediction['row_id'].min())
    print("Min prediction_unit_id : ",test['prediction_unit_id'].min())
    print("Date max : ", test['prediction_datetime'].max())
    print("Max prediction_unit_id : ",test['prediction_unit_id'].max())
    print("Row_id max test: ", test['row_id'].max())
    print("Row_id max sample prediction : ", sample_prediction['row_id'].max())
    print("Nb uniques prediction_unit_id : ",test['prediction_unit_id'].nunique())
    print("Nb uniques test : ",test['row_id'].nunique())
    print("Nb uniques sample prediction : ",sample_prediction['row_id'].nunique())
    print(" ======== STOP ======== ")
    print("Date forecast elec : ", electricity_prices['forecast_date'].min())
    print("Date max forecast elec : ", electricity_prices['forecast_date'].max())
    print("Date forecast gas : ", gas_prices['forecast_date'].min())
    print("Date max forecast gas : ", gas_prices['forecast_date'].max())
    type(electricity_prices['forecast_date'][0])
    type(gas_prices['forecast_date'][0])
    
    
    sample_prediction['target'] = 0
    env.predict(sample_prediction)
    counter+=1
```
